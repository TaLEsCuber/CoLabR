```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

def draw_energy_levels():
    fig, ax = plt.subplots(figsize=(10, 8))

    # Define energy levels and their positions
    y_base = {
        "5S1/2": 0,
        "5P1/2": 3,
        "5P3/2": 6,
    }

    # Hyperfine levels for Rb-85 and Rb-87
    hf_levels = {
        "Rb85": {
            "5S1/2": [2, 3],
            "5P1/2": [2, 3],
            "5P3/2": [1, 2, 3, 4],
        },
        "Rb87": {
            "5S1/2": [1, 2],
            "5P1/2": [1, 2],
            "5P3/2": [0, 1, 2, 3],
        }
    }

    # Zeeman splitting visualization: offset for mF sublevels (not real energies)
    mF_offsets = [-0.2, -0.1, 0, 0.1, 0.2, 0.3]

    colors = {
        "Rb85": "tab:blue",
        "Rb87": "tab:red"
    }

    for isotope, levels in hf_levels.items():
        color = colors[isotope]
        for state, F_levels in levels.items():
            y = y_base[state]
            for F in F_levels:
                x_center = y + (0.2 if isotope == "Rb85" else -0.2)
                # Draw central F level
                ax.hlines(y + 0.1 * F, x_center - 0.1, x_center + 0.1, color=color, linewidth=2)
                # Draw Zeeman sublevels
                for i, offset in enumerate(mF_offsets[:2 * F + 1]):
                    ax.hlines(y + 0.1 * F + offset * 0.3, x_center - 0.05, x_center + 0.05, color=color, alpha=0.5)

                ax.text(x_center + 0.15, y + 0.1 * F, f'F={F}', color=color, fontsize=8, verticalalignment='center')

    # Labels
    for state, y in y_base.items():
        ax.text(0, y + 0.5, state, fontsize=10, fontweight='bold', verticalalignment='center')

    # Legends
    rb85_patch = mpatches.Patch(color='tab:blue', label='Rb-85')
    rb87_patch = mpatches.Patch(color='tab:red', label='Rb-87')
    ax.legend(handles=[rb85_patch, rb87_patch], loc='upper left')

    ax.set_xlim(-0.5, 7)
    ax.set_ylim(-1, 8)
    ax.axis('off')
    ax.set_title("Rb-85 and Rb-87 Energy Levels with Fine, Hyperfine, and Zeeman Structure", fontsize=12)

    plt.tight_layout()
    plt.show()

draw_energy_levels()
```

```python
import matplotlib.pyplot as plt

def draw_energy_diagram(title, isotope, F_levels_5S, F_levels_5P1_2, F_levels_5P3_2):
    fig, ax = plt.subplots(figsize=(10, 8))

    # 三层结构高度设定（超精细结构到精细结构的gap，精细结构到主能级的gap）
    hyperfine_gap = 3   # 超精细结构内部能级间距
    fine_gap = 6        # 精细结构间距
    coarse_gap = 12     # 主能级间距

    zeeman_dx = 0.13    # 水平方向偏移（mF用）

    color = 'tab:blue'
    fine_structure_color = 'dimgray'  # 精细结构线颜色
    coarse_structure_color = 'black'  # 主能级线颜色

    # 计算超精细能级的y坐标集合
    def get_hyperfine_ys(F_levels, base_y):
        return [base_y + i * hyperfine_gap for i in range(len(F_levels))]

    # 精细结构y坐标，是对应超精细结构y范围的中点
    y_5S_hyperfine = 0
    y_5S_fine = y_5S_hyperfine + (len(F_levels_5S)-1)*hyperfine_gap / 2 + fine_gap

    y_5P1_2_hyperfine = y_5S_fine + coarse_gap
    y_5P1_2_fine = y_5P1_2_hyperfine + (len(F_levels_5P1_2)-1)*hyperfine_gap / 2 + fine_gap

    y_5P3_2_hyperfine = y_5P1_2_fine + coarse_gap
    y_5P3_2_fine = y_5P3_2_hyperfine + (len(F_levels_5P3_2)-1)*hyperfine_gap / 2 + fine_gap

    # 画主能级线（黑色粗线）
    ax.hlines(y_5S_fine + coarse_gap, 0.2, 0.4, color=coarse_structure_color, linewidth=5, alpha=0.8)
    ax.hlines(y_5P1_2_fine + coarse_gap, 0.2, 0.4, color=coarse_structure_color, linewidth=5, alpha=0.8)
    ax.hlines(y_5P3_2_fine + coarse_gap, 0.2, 0.4, color=coarse_structure_color, linewidth=5, alpha=0.8)

    # 画精细结构线（灰色较粗线）
    ax.hlines(y_5S_fine, 0.5, 0.8, color=fine_structure_color, linewidth=4, alpha=0.7)
    ax.hlines(y_5P1_2_fine, 0.5, 0.8, color=fine_structure_color, linewidth=4, alpha=0.7)
    ax.hlines(y_5P3_2_fine, 0.5, 0.8, color=fine_structure_color, linewidth=4, alpha=0.7)

    # 画主能级到精细结构的垂直连接线
    ax.vlines(0.35, y_5S_fine, y_5S_fine + coarse_gap, color='black', linewidth=1)
    ax.vlines(0.35, y_5P1_2_fine, y_5P1_2_fine + coarse_gap, color='black', linewidth=1)
    ax.vlines(0.35, y_5P3_2_fine, y_5P3_2_fine + coarse_gap, color='black', linewidth=1)

    # 画精细结构到超精细结构的垂直连接线
    ax.vlines(0.9, y_5S_hyperfine, y_5S_fine, color=fine_structure_color, linewidth=1)
    ax.vlines(0.9, y_5P1_2_hyperfine, y_5P1_2_fine, color=fine_structure_color, linewidth=1)
    ax.vlines(0.9, y_5P3_2_hyperfine, y_5P3_2_fine, color=fine_structure_color, linewidth=1)

    # 画超精细结构和Zeeman分裂
    def draw_levels(F_levels, base_y):
        for i, F in enumerate(F_levels):
            y = base_y + i * hyperfine_gap
            # 超精细能级线（不偏移）
            ax.hlines(y, 1.0, 1.4, color=color, linewidth=2)
            ax.text(0.9, y, f"F={F}", verticalalignment='center', fontsize=10)
            # Zeeman splitting（带水平偏移）
            for mF in range(-F, F + 1):
                xm = 2 + mF * zeeman_dx
                ym = y + mF * 0.5
                ax.hlines(ym, xm - 0.05, xm + 0.05, color=color, linewidth=1, alpha=0.6)
                # 标注 m_F
                ax.text(xm + 0.06, ym, f"$m_F = {mF}$", verticalalignment='center', fontsize=8, color=color, alpha=0.5)

    draw_levels(F_levels_5S, y_5S_hyperfine)
    draw_levels(F_levels_5P1_2, y_5P1_2_hyperfine)
    draw_levels(F_levels_5P3_2, y_5P3_2_hyperfine)

    # 标签
    ax.text(0.1, y_5S_fine + coarse_gap + 0.3, "5S (Coarse)", fontsize=12)
    ax.text(0.1, y_5S_fine + 0.3, "5S$_{1/2}$ (Fine)", fontsize=12)
    ax.text(0.1, y_5S_hyperfine - 0.5, "5S Hyperfine", fontsize=12, color=color)

    ax.text(0.1, y_5P1_2_fine + coarse_gap + 0.3, "5P$_{1/2}$ (Coarse)", fontsize=12)
    ax.text(0.1, y_5P1_2_fine + 0.3, "5P$_{1/2}$ (Fine)", fontsize=12)
    ax.text(0.1, y_5P1_2_hyperfine - 0.5, "5P$_{1/2}$ Hyperfine", fontsize=12, color=color)

    ax.text(0.1, y_5P3_2_fine + coarse_gap + 0.3, "5P$_{3/2}$ (Coarse)", fontsize=12)
    ax.text(0.1, y_5P3_2_fine + 0.3, "5P$_{3/2}$ (Fine)", fontsize=12)
    ax.text(0.1, y_5P3_2_hyperfine - 0.5, "5P$_{3/2}$ Hyperfine", fontsize=12, color=color)

    ax.set_xlim(0, 3)
    ax.set_ylim(-2, y_5P3_2_fine + coarse_gap + 5)
    ax.axis('off')
    ax.set_title(f"{isotope} Hyperfine Structure with Fine and Coarse Splittings", fontsize=16)
    plt.tight_layout()
    plt.show()


# Rb-87 数据
F_5S_Rb87 = [1, 2]
F_5P1_2_Rb87 = [1, 2]
F_5P3_2_Rb87 = [0, 1, 2, 3]

# Rb-85 数据
F_5S_Rb85 = [2, 3]
F_5P1_2_Rb85 = [2, 3]
F_5P3_2_Rb85 = [1, 2, 3, 4]

# 分别绘图
draw_energy_diagram("Rb-87", "Rb-87", F_5S_Rb87, F_5P1_2_Rb87, F_5P3_2_Rb87)
draw_energy_diagram("Rb-85", "Rb-85", F_5S_Rb85, F_5P1_2_Rb85, F_5P3_2_Rb85)
```
