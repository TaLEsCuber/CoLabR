```python
import numpy as np
import matplotlib.pyplot as plt

# 
def epsilon(z, delta, a):
    return 1 + delta - delta / (np.exp(z / a) + 1)

# 
z_min, z_max = -10.0, 10.0  # 
dz = 0.01                 # 
z = np.arange(z_min, z_max, dz)

# 
delta = 0.1
a = 0.5

eps_values = epsilon(z, delta, a)

plt.figure(dpi=300)
plt.plot(z, eps_values, label=f'delta={delta}, a={a}')
plt.xlabel('z')
plt.ylabel('epsilon(z)')
plt.title('Variation of Dielectric Constant')
plt.legend()
plt.grid(True)
plt.show()
```
