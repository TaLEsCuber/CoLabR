```python
# 导入必要的库
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.style.use('science')
```

```python
# 输入数据
x = np.array([20, 35, 70, 100, 200, 500, 700, 1000])
y = np.array([187.4, 271.9, 351.9, 374.4, 393.6, 399.9, 400.6, 400.9]) / 50.0

# 拟合函数
def fit_func(x, c3, r1, r3):
    return 4 * np.pi * x * c3 * r1 / (np.sqrt(1 + (2 * np.pi * c3 * r3)**2))

# 进行拟合
initial_guess = [1e-6, 20e3, 4.7e3]
popt, pcov = curve_fit(fit_func, x, y, p0=initial_guess)

# 生成拟合曲线的x值
x_fit = np.linspace(min(x), max(x), 1000)

# 计算拟合曲线的y值
y_fit = fit_func(x_fit, *popt)

# 绘制原始数据和拟合曲线
plt.figure(dpi=500)
plt.scatter(x, y, label='Original data')
plt.plot(x_fit, y_fit, 'r', label='Fitted curve')
plt.xlabel('f/Hz')
plt.ylabel('Av')
plt.title('Av-f Relation')
plt.show()

# 输出拟合参数
print("拟合参数 (p1):", popt)
```
